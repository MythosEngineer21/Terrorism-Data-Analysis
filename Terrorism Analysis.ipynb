### Terror Attack City

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas dataframe
df = pd.read_csv('terrorismData.csv', encoding='ISO-8859-1')

# Filter the dataframe to include only attacks in Jammu and Kashmir
jk_df = df[df['State'] == 'Jammu and Kashmir']

# Group the attacks by city and count the number of attacks
city_counts = jk_df['City'].value_counts()

# Get the most dangerous city (i.e., the city with the most attacks)
most_dangerous_city = city_counts.index[0]

# Filter the dataframe to include only attacks in the most dangerous city
most_dangerous_city_df = jk_df[jk_df['City'] == most_dangerous_city]

# Filter out attacks with unknown terrorist groups
most_dangerous_city_df = most_dangerous_city_df[most_dangerous_city_df['Group'] != 'Unknown']

# Count the number of attacks by each terrorist group and get the most active group
group_counts = most_dangerous_city_df['Group'].value_counts()
most_active_group = group_counts.index[0]

# Print the results
print(most_dangerous_city, city_counts[0],  most_active_group)

# Get the top 5 cities with most attacks
top_cities = city_counts.head(5)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Attacks')
plt.title('Top 5 Cities with Most Terror Attacks in Jammu and Kashmir')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Terror Frequency

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("terrorismData.csv", skipinitialspace=True)

# Counting for the Red States
red_states = ['Jharkhand', 'Odisha', 'Andhra Pradesh', 'Chhattisgarh']
red_data = df[df['State'].isin(red_states)]
n = df['Year'].nunique()
red_data['Casualty'] = red_data['Killed'] + red_data['Wounded']
freq_1 = int(red_data['Casualty'].sum() / n)

# Counting for Jammu and Kashmir
jk_data = df[df['State'] == 'Jammu and Kashmir']
jk_data['Casualty'] = jk_data['Killed'] + jk_data['Wounded']
freq_2 = int(jk_data['Casualty'].sum() / n)

# Printing the results
print('{} {}'.format(freq_1, freq_2))


# Plotting Average Casualties

# Calculate average casualties per year
red_avg_casualties = red_data.groupby('Year')[['Killed', 'Wounded']].sum().mean(axis=1)
jk_avg_casualties = jk_data.groupby('Year')[['Killed', 'Wounded']].sum().mean(axis=1)

# Create a grouped bar chart
plt.figure(figsize=(10, 6))
plt.bar(red_avg_casualties.index, red_avg_casualties, width=0.4, align='center', label='Red States')
plt.bar(jk_avg_casualties.index + 0.4, jk_avg_casualties, width=0.4, align='center', label='Jammu and Kashmir')
plt.xlabel('Year')
plt.ylabel('Average Casualties')
plt.title('Average Casualties in Red States vs Jammu and Kashmir')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### Temporal Trends

```python
# Load the dataset
df = pd.read_csv("terrorismData.csv", encoding='ISO-8859-1')

# --------------------------
# Insight 1: Temporal Trends
# --------------------------

# Group data by year and count attacks
attacks_per_year = df.groupby('Year')['Year'].count()

# Create a line chart
plt.figure(figsize=(10, 6))
attacks_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Terror Attacks Trend (1970-2017)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Explanation
print("Insight 1: Temporal Trends")
print("The line chart displays the trend of terrorist attacks over the years (1970-2017).")
print("This visualization helps us understand if there's an upward or downward trend in attacks.")
print("---------------------------------------\n")
```

### Geographical Patterns

```python
# --------------------------
# Insight 2: Geographical Patterns
# --------------------------

# Group data by country and count attacks
country_counts = df['Country'].value_counts().head(10)

# Create a bar chart for top 10 countries
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Countries with Most Terror Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Explanation
print("Insight 2: Geographical Patterns")
print("The bar chart presents the top 10 countries with the highest number of terrorist attacks.")
print("This visualization helps identify areas that are more susceptible to terrorism.")
print("---------------------------------------\n")
```

### Attack Types

```python
# --------------------------
# Insight 3: Attack Types
# --------------------------

# Group data by attack type and count attacks
attack_type_counts = df['AttackType'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(attack_type_counts, labels=attack_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Terrorist Attack Types')
plt.axis('equal')
plt.show()

# Explanation
print("Insight 3: Attack Types")
print("The pie chart illustrates the distribution of different types of terrorist attacks.")
print("This visualization helps us understand the preferred methods of terrorists.")
print("---------------------------------------\n")
```

### Casualties and Impact

```python
# --------------------------
# Insight 4: Casualties and Impact
# --------------------------

# Calculate total casualties (killed + wounded)
df['TotalCasualties'] = df['Killed'] + df['Wounded']

# Group data by year and calculate total casualties
casualties_per_year = df.groupby('Year')['TotalCasualties'].sum()

# Create a line chart
plt.figure(figsize=(10, 6))
casualties_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Casualties')
plt.title('Total Casualties Trend (1970-2017)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Explanation
print("Insight 4: Casualties and Impact")
print("The line chart displays the trend of total casualties (killed + wounded) over the years.")
print("This visualization helps us understand the impact and severity of attacks.")
print("---------------------------------------\n")
```
